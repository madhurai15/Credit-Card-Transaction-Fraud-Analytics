```python
num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
```

```python
df['trans_date_trans_time'].dtype
```

```python
df = df[df['amt'] > 0]
```

```python
df['Month'] = df['trans_date_trans_time'].dt.month_name()
```

```python
df['Day_of_Week'] = df['trans_date_trans_time'].dt.day_name()
```

```python
df['Weekend'] = np.where(df['Day_of_Week'].isin(['Saturday','Sunday']), 'Yes', 'No')
```

```python
df['dob'] = pd.to_datetime(df['dob'])

df['Age'] = (df['trans_date_trans_time'] - df['dob']).dt.days // 365
```

```python
bins = [0,18,25,35,45,55,65,100]

labels = ['<18','18-25','26-35','36-45','46-55','56-65','65+']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
category_amount = df.groupby('category')['amt'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))

category_amount.plot(kind='bar')

plt.title("Transaction Amount by Category")
plt.xlabel("Category")
plt.ylabel("Total Amount")

plt.xticks(rotation=45)

plt.show()
```

```python
city_amount = df.groupby('city')['amt'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))

city_amount.plot(kind='bar')

plt.title("Top 10 Cities by Spending")
plt.xlabel("City")
plt.ylabel("Total Spending")

plt.xticks(rotation=45)

plt.show()
```

```python
age_group = df.groupby('Age_Group')['amt'].sum()

plt.figure(figsize=(8,5))

age_group.plot(kind='bar')

plt.title("Spending by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Spending")

plt.show()
```

```python
gender = df.groupby('gender')['amt'].sum()

plt.figure(figsize=(6,5))

gender.plot(kind='bar')

plt.title("Spending by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Spending")

plt.show()
```

```python
month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly = df.groupby('Month')['amt'].sum().reindex(month_order)

plt.figure(figsize=(12,5))

plt.plot(monthly.index, monthly.values, marker='o')

plt.title("Monthly Spending Trend")
plt.xlabel("Month")
plt.ylabel("Total Spending")

plt.xticks(rotation=45)

plt.show()
```

```python
day_order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

day = df.groupby('Day_of_Week')['amt'].sum().reindex(day_order)

plt.figure(figsize=(10,5))

plt.plot(day.index, day.values, marker='o')

plt.title("Day-wise Spending")
plt.xlabel("Day")
plt.ylabel("Total Spending")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.boxplot(df['amt'])

plt.title("Transaction Amount Box Plot")
plt.ylabel("Amount")

plt.show()
```

```python
fraud = df['is_fraud'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(fraud,
        labels=['Non-Fraud','Fraud'],
        autopct='%1.1f%%',
        startangle=90)

plt.title("Fraud Distribution")

plt.show()
```
